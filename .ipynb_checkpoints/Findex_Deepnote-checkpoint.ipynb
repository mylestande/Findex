# Findex 2021

Financial inclusion is a cornerstone of development, and since 2011, the Global Findex Database has been the definitive source of data on global access to financial services from payments to savings and borrowing. The 2021 edition, based on nationally representative surveys of about 128,000 adults in 123 economies during the COVID-19 pandemic, contains updated indicators on access to and use of formal and informal financial services and digital payments, and offers insights into the behaviours that enable financial resilience. The data also identify gaps in access to and usage of financial services by women and poor adults.

We will use the country-level data from 2021 (it is up to 2021 actually, including the data from 2011, 2014, 2017, and 2021). The data and codebook are publicly accessible here and are also available on Learn: https://www.worldbank.org/en/publication/globalfindex/Data 

The data is given in an Excel file, the first sheet includes the country-level data and the second sheet is the codebook describing the variables.

```python

# Convert percentage strings to float for the specified columns
columns_to_convert = [
    'Made or received a digital payment, male (% age 15+)',
    'Made or received a digital payment, female (% age 15+)',
    'Mobile money account (% age 15+)',
    'Account, male (% age 15+)',
    'Account, female (% age 15+)',
    'Received government payments, male (% age 15+)',
    'Received government payments, female (% age 15+)',
    'Paid school fees, male (% age 15+)',
    'Paid school fees, female (% age 15+)',
    'Sent or received domestic remittances, male (% age 15+)',
    'Sent or received domestic remittances, female (% age 15+)',
    'Saved any money, male (% age 15+)',
    'Saved any money, female (% age 15+)',
    'Mobile money account, male (% age 15+)',
    'Mobile money account, female (% age 15+)',
    'Received wages, male (% age 15+)',
    'Received wages, female (% age 15+)',
    'Coming up with emergency funds in 7 days: possible, male (% age 15+)',
    'Coming up with emergency funds in 7 days: possible, female (% age 15+)',
    'Borrowed any money, male (% age 15+)',
    'Borrowed any money, female (% age 15+)'
]

# Remove percentage sign and convert to float
#for column in columns_to_convert:
#    data[column] = data[column].str.rstrip('%').astype('float')

# Calculate gender gaps and other metrics
data['DigitalPaymentGenderGap'] = data['Made or received a digital payment, male (% age 15+)'] - data['Made or received a digital payment, female (% age 15+)']
data['MobileMoney'] = data['Mobile money account (% age 15+)']
data['AccountOwnershipGenderGap'] = data['Account, male (% age 15+)'] - data['Account, female (% age 15+)']
data['ReceivedGovtPaymentsGenderGap'] = data['Received government payments, male (% age 15+)'] - data['Received government payments, female (% age 15+)']
data['PaySchoolFeesGenderGap'] = data['Paid school fees, male (% age 15+)'] - data['Paid school fees, female (% age 15+)']
data['OverallAccountGenderGap'] = data['Account, male (% age 15+)'] - data['Account, female (% age 15+)']
data['SendReceiveRemittanceGenderGap'] = data['Sent or received domestic remittances, male (% age 15+)'] - data['Sent or received domestic remittances, female (% age 15+)']
data['AnySavingsGenderGap'] = data['Saved any money, male (% age 15+)'] - data['Saved any money, female (% age 15+)']
data['MobileMoneyGenderGap'] = data['Mobile money account, male (% age 15+)'] - data['Mobile money account, female (% age 15+)']
data['ReceiveWagesGenderGap'] = data['Received wages, male (% age 15+)'] - data['Received wages, female (% age 15+)']
data['EmergencyFund7DaysGenderGap'] = data['Coming up with emergency funds in 7 days: possible, male (% age 15+)'] - data['Coming up with emergency funds in 7 days: possible, female (% age 15+)']
data['BorrowedAnyMoneyGenderGap'] = data['Borrowed any money, male (% age 15+)'] - data['Borrowed any money, female (% age 15+)']

data.head()
```

```python
selected_cols = [
    "Country name",
    "Country code",
    "Year",
    "Adult populaiton",
    "Region",
    "Income group",
    "MobileMoney",
    "DigitalPaymentGenderGap",
    "AccountOwnershipGenderGap",
    "ReceivedGovtPaymentsGenderGap",
    "PaySchoolFeesGenderGap",
    "OverallAccountGenderGap",
    "SendReceiveRemittanceGenderGap",
    "AnySavingsGenderGap",
    "MobileMoneyGenderGap",
    "ReceiveWagesGenderGap",
    "EmergencyFund7DaysGenderGap",
    "BorrowedAnyMoneyGenderGap"
]

workingdata = data[selected_cols].copy()
workingdata
```

```python
from statsmodels.formula.api import ols

reg_model = ols(
    "DigitalPaymentGenderGap ~ Region + Q('Income group') + ReceivedGovtPaymentsGenderGap + OverallAccountGenderGap + AnySavingsGenderGap + ReceiveWagesGenderGap + EmergencyFund7DaysGenderGap + BorrowedAnyMoneyGenderGap",
    data=workingdata_2021,
).fit()
reg_model.summary()
```

```python
fitted_values = reg_model.fittedvalues
residuals = reg_model.resid

plt.figure(figsize=(12, 8))
sns.scatterplot(x=fitted_values, y=residuals)
plt.axhline(y=0, color='red', linestyle='--')
plt.axvline(x=0, color='red', linestyle='--')
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals (Errors)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.histplot(residuals, kde=True, bins=50)
plt.title('Histogram of Residuals')
plt.xlabel('Residual Value')
plt.ylabel('Frequency')
plt.show()
```

```python
from statsmodels.api import qqplot
qqplot(data=residuals, fit=True, line='s')
plt.title('Q-Q Plot of Residuals')
plt.show()
```

```python
print("R-squared: ", reg_model.rsquared)
print("RMSE: ", np.sqrt(reg_model.mse_resid))
print("Adjusted R-Squared: ", reg_model.rsquared_adj)
```

```python
# print(data16.dtypes)
# print(data16.select_dtypes(exclude=['object', 'category']).dtypes)
reg_model_2 = ols(
    "DigitalPaymentGenderGap ~ ReceivedGovtPaymentsGenderGap + OverallAccountGenderGap + AnySavingsGenderGap + ReceiveWagesGenderGap + EmergencyFund7DaysGenderGap + BorrowedAnyMoneyGenderGap",
    data=workingdata_2021,
).fit()

#reg_model_2.params
#reg_model_2.summary()

reg_2_mean_values = workingdata_2021[["ReceivedGovtPaymentsGenderGap", "OverallAccountGenderGap", "AnySavingsGenderGap", "ReceiveWagesGenderGap", "EmergencyFund7DaysGenderGap", "BorrowedAnyMoneyGenderGap"]].mean()
predicted_adr = reg_model_2.predict(reg_2_mean_values)

print("The DigitalPaymentGenderGap prediction for the means of the numerical predictors is: ", predicted_adr.item())

# raise NotImplementedError()
```
